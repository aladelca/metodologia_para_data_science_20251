```python
# Importar las bibliotecas necesarias
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
# Definir los símbolos de las acciones
tickers = ['DIS', 'AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA', 'TSLA', 'META', 'BRK-B', 'AVGO']
```

```python
# Rango de fechas: últimos 12 meses hasta hoy
end_date = datetime.today().strftime('%Y-%m-%d')
start_date = (datetime.today() - timedelta(days=365)).strftime('%Y-%m-%d')
```

```python
# Descargar los datos desde Yahoo Finance
data = yf.download(tickers, start=start_date, end=end_date)
```

```python
# Extraer solo precios de cierre
close_prices = data['Close']
```

```python
# Mostrar los primeros registros
close_prices.head()
```

```python
# Crear el gráfico de líneas
plt.figure(figsize=(16, 8))
for ticker in tickers:
    plt.plot(close_prices.index, close_prices[ticker], label=ticker)

# Personalización del gráfico
plt.title('Evolución del Precio de Cierre de Acciones (Últimos 12 meses)')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre (USD)')
plt.legend(loc='upper left', fontsize=9)
plt.grid(True)
plt.tight_layout()
plt.show()
```
